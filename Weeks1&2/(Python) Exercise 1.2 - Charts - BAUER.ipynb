# Bar Chart

```python
plt.figure(figsize=(20,10)) 
plt.bar(hotdog['Year'], hotdog['Dogs eaten'], color = 'green')
plt.xlabel("Year", fontsize=15)
plt.ylabel("Number of Hotdogs Eaten by Winner", fontsize=15)
plt.title("Number of Hotdogs Eaten by the Contest Winner by Year", fontsize=18)
plt.show()
```

# Stacked Bar Chart

```python
# horizontal stacked bar chart
# source for help: https://towardsdatascience.com/stacked-bar-charts-with-pythons-matplotlib-f4020e4eb4a7
fields = ['Approve','Disapprove','None']
colors = ['#1D2F6F', '#8390FA', '#6EAF46']
labels = ['Approve','Disapprove','None']

# figure and axis
fig, ax = plt.subplots(1, figsize=(12, 10))

# plot bars
left = len(obama) * [0]
for idx, name in enumerate(fields):
    plt.barh(obama['Issue'], obama[name], left = left, color=colors[idx])
    left = left + obama[name]

# title, legend, labels
plt.title('Obama\'s Approval Ratings By Issue', loc='left', fontweight='bold')
plt.legend(labels, bbox_to_anchor=([0.55, 1, 0, 0]), ncol=4, frameon=False)
plt.xlabel('Percentage')

# adjust limits and draw grid lines
plt.ylim(-0.5, ax.get_yticks()[-1] + 0.5)
ax.set_axisbelow(True)
ax.xaxis.grid(color='gray', linestyle='dashed')
plt.show()
```

# Pie Chart

```python
# figure size
plt.figure(figsize=(8,8)) 

# set up
labels = obama['Issue']
approve = obama['Approve']

# construct pie chart
plt.pie(approve, labels = labels, autopct='%1.1f%%')
plt.axis('equal')
plt.suptitle('Obama\'s Approval Rating by Issue', fontsize=18, fontweight='bold')
plt.show()
```

# Donut Chart

```python
#Donut graph (extension of pie chart code)

# figure size
plt.figure(figsize=(8,8)) 

# set up
labels = obama['Issue']
approve = obama['Approve']

# construct pie chart
plt.pie(approve, labels = labels,
        autopct='%1.1f%%', shadow=True)
        
# Draw a circle at the center of pie to make it look like a Donut
centre_circle = plt.Circle((0,0),0.75,color='black', fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.axis('equal')
plt.suptitle('Obama\'s Approval Rating by Issue', fontsize=18, fontweight='bold')
plt.show()
```
